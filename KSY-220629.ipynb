```python
# 학생의 이름과 점수(국어, 영어, 수학)를 입력 받아 저장한 후
# 저장된 자료의 이름과 각 학생의 성적, 총점, 평균을 구하세요
# 가장 성적이 좋은 학생의 이름과 총점을 출력
# 가장 성적이 나쁜 학생의 이름과 총점을 출력
# 전체 학생의 인원수와 전체 평균을 구하세요
# 조건) class사용, 모든 문제는 함수로 구현
```

```python
class Student:
    num_students = 0
    students = []
    
    @classmethod
    def print(cls):
        print('------- 학생 목록 -------')
        print("이름\t국어\t영어\t수학\t총점\t평균")
#         for student in Student.students 도 가능
        for student in cls.students:
            print(str(student))
        print('-------------------------')
        
    def search_Student(cls,value):
        for student in cls.students:
            if student.get_name() == value.name:
                print(student)
            else:
                print("찾는 학생이 없습니다.")
                  
            
    def __init__(self, name, kor, eng, math):
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = math
        
        Student.num_students += 1
        Student.students.append(self)
    
    def __str__(self):
        return "{}\t{}\t{}\t{}\t{}\t{:3.2f}".format(self.name, 
                                                    self.kor, self.eng, self.math,
                                                    self.get_sum(), self.get_avg())
    
    def get_sum(self):
        return self.kor + self.eng + self.math
    
    def get_avg(self):
        return self.get_sum() / 3
    
    def get_name(self):
        return self.name
    
    def __gt__(self, value):
        return self.get_sum() > value.get_sum()
    
def input_data():
    while True:
        name = input("이름을 입력하세요 : ")
        
        if name == 'q':
            break
        
        while True:
            grade = input("국어 영어 수학 성적 입력 : ").split()

            if len(grade) != 3:
                print("다시 입력하세요.")
                print("ex)10 20 30")
                continue
            else:
                break
        kor = int(grade[0])
        eng = int(grade[1])
        math = int(grade[2])
        
        Student(name, kor, eng, math)

def menu():
    while True:
        print("1.학생 검색")
        print("2.가장 성적이 좋은 학생")
        print("3.가장 성적이 나쁜 학생")
        print("4.총 학생수와 전체 평균")
        print("=========================")
        choice = input("입력 : ")
        
        if choice in '1234':
            if choice == '1':
                search_name = input("검색할 학생의 이름 : ")
                Student.search_Student(Student, search_name)
                continue
        elif choice in 'qQ':
            break
        else:
            print("다시 입력하세요.")
            print("=========================")
            continue
        
```

```python
input_data()
Student.print()
```
